# AND pattern using Perceptron

```python
import numpy as np
```

```python
# define activation function
def step_function(x):
    return 1 if x>=0 else 0
```

```python
# perceptron algorithm
def perceptron(x,y, lr=0.1, epochs=10):
    weights = np.zeros(x.shape[1])
    bias = 0

    for epoch in range(epochs):
        for i in range(len(x)):
            linear_output = np.dot(x[i], weights)+bias
            prediction = step_function(linear_output)
            error = y[i] - prediction
            weights+= lr*error*x[i]
            bias+= lr*error
    return weights, bias
```

```python
# input data for AND gate
x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0,0,0,1])
weights, bias = perceptron(x,y)
```

```python
def predict(x, weights, bias):
    linear_output = np.dot(x, weights) + bias
    return step_function(linear_output)

for xs in x:
    print(f"Input : {xs}, Predicted Output {predict(xs, weights, bias)}")
```

# Match the given input patterns

```python
import re
import pandas as pd

titanic = pd.read_csv('/Users/sohan/Downloads/Titanic-Dataset.csv')

print("Dataset columns:")
print(titanic.columns)

def match_pattern(data, column, pattern):
    results = []
    for value in data[column]:
        if re.match(pattern, str(value)):
            results.append(value)
    return results

def user_input():
    print("\nAvailable columns to search:")
    for col in titanic.columns:
        print(f" - {col}")
    
    column = input("Enter the column to search for: ").strip()

    if column not in titanic.columns:
        print("Invalid column name. Please select a valid column.")
        return

    pattern = input(f"Enter the pattern to search in the '{column}' column: ").strip()
    results = match_pattern(titanic, column, pattern)

    if results:
        print(f"\nFound {len(results)} matches for the pattern '{pattern}' in column '{column}':")
        for result in results:
            print(result)
    else:
        print(f"\nNo match found for the pattern '{pattern}'.")

def main():
    print("Welcome to the Titanic dataset pattern matching program!")
    while True:
        user_input()
        again = input("\nDo you want to search again? (yes/no): ").strip().lower()
        if again != 'yes':
            print("Goodbye!")
            break

main()
```
